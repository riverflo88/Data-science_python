```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
insurance = pd.read_csv("insurance.csv")
insurance.head()
```

```python
#getting age and charges data as x and y resp.
x = insurance.iloc[:,0].values.reshape(-1, 1)
y = insurance.iloc[:,-1].values
print(x.shape,y.shape)
```

```python
#
plt.scatter(x,y,color='green')
plt.title("Age Vs Insurance cost")
plt.xlabel("Age")
plt.ylabel("Insurance Cost");
plt.show()
```

```python
#instatiate the model
age_model = LinearRegression()
#fit the model
age_model.fit(x,y)
```

```python
#predict
y_predict = age_model.predict(x)
```

```python
#scatter plot plus best fit
plt.scatter(x,y,color='green')
plt.plot(x,y_predict,color="red")
plt.title("Age Vs Insurance cost")
plt.xlabel("Age")
plt.ylabel("Insurance Cost");
plt.show()
```
